```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.datasets import make_circles
from sklearn.cluster import DBSCAN
# from DBScanpp import DBScanpp
from sklearn.metrics import accuracy_score

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
import warnings
warnings.filterwarnings("ignore")
```

```python
def grafic(x,y,color):
    plt.scatter(x=x, y=y, c=color)
    plt.xlabel('Eje X')
    plt.ylabel('Eje Y')
    plt.title('Gráfico de dispersión con etiquetas de clúster')
    plt.show()
```

```python
X, y = make_circles(
    # -------------------------------------------------------------------------
    # The number of samples.
    n_samples=10000,
    # -------------------------------------------------------------------------
    # Shuffle the samples.
    shuffle=False,
    # -------------------------------------------------------------------------
    # Standard deviation of Gaussian noise added to the data.
    noise=0.05,
    # -------------------------------------------------------------------------
    # Determines random number generation for dataset shuffling and noise. Pass
    # an int for reproducible output across multiple function calls.
    # creation.
    random_state=420,
    # -------------------------------------------------------------------------
    # Scale factor between inner and outer circle in the range (0, 1).
    factor=0.8,
)
```

## Probar con DBScan

```python
dbscan = DBSCAN(eps=0.06, 
                min_samples=50)
clusters = dbscan.fit(X)
```

```python
print(accuracy_score(clusters.labels_, y))
set(clusters.labels_)
```

```python
grafic(x=X[:, 0], y=X[:, 1], color=clusters.labels_)
```
