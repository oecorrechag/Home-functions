```python
creditos = pd.DataFrame({
    "Cliente": ["Juan", "Juan", "Ana", "Ana", "Pedro", "Pedro"],
    "Monto": [1000, 5000, 2000, 3000, 1500, 2000],
    "Cuotas": [12, 24, 6, 12, 8, 12],
    "Fecha": ["2022-01-01", "2022-02-01", "2022-01-01", "2022-03-01", "2022-01-01", "2022-02-01"],
    "dias_mora":[0,0,25,0,11,0]
})
creditos["Fecha"] = pd.to_datetime(creditos["Fecha"])
creditos.head()
```

```python
creditos = creditos.sort_values(by=['Fecha'], ascending=False)
creditos = creditos.groupby('Cliente').head(1).reset_index()
creditos['Antiguedad'] = (dt.datetime.now() - creditos['Fecha']).dt.days
creditos
```

## 2. Ciclo de vida

```python
creditos = pd.DataFrame({
    "Cliente": ["Juan", "Juan", "Ana", "Ana", "Pedro", "Pedro"],
    "Monto": [1000, 5000, 2000, 3000, 1500, 2000],
    "Cuotas": [12, 24, 6, 12, 8, 12],
    "Fecha": ["2022-01-01", "2022-02-01", "2022-01-01", "2022-03-01", "2022-01-01", "2022-02-01"],
    "dias_mora":[0,0,25,0,11,0]
})
creditos["Fecha"] = pd.to_datetime(creditos["Fecha"])
creditos.head()
```

```python
# crear una nueva columna con la fecha de inicio del ciclo de vida de cada cliente
fecha_inicio = creditos.groupby('Cliente')['Fecha'].min()
fecha_inicio
```

```python
creditos['inicio_ciclo_vida'] = creditos['Cliente'].map(fecha_inicio)
creditos
```

```python
# crear una nueva columna con la fecha de fin del ciclo de vida de cada cliente
fecha_fin = creditos.groupby('Cliente')['Fecha'].max()
fecha_fin
```

```python
creditos['fin_ciclo_vida'] = creditos['Cliente'].map(fecha_fin)
creditos
```

```python
# calcular la duración del ciclo de vida de cada cliente
creditos['ciclo_vida'] = (pd.to_datetime(creditos['fin_ciclo_vida']) - pd.to_datetime(creditos['inicio_ciclo_vida'])).dt.days
creditos
```

```python
creditos = creditos.groupby('Cliente').head(1).reset_index()
creditos
```

## 3. Tiempo promedio y mediana entre creditos

```python
creditos = pd.DataFrame({
    "Cliente": ["Juan", "Juan", "Ana", "Ana", "Pedro", "Pedro","Ana","Ana"],
    "Monto": [1000, 5000, 2000, 3000, 1500, 2000, 5000, 300],
    "Cuotas": [12, 24, 6, 12, 8, 12, 24, 2],
    "Fecha": ["2022-01-01", "2022-02-01", "2022-01-01", "2022-03-01", "2022-01-01", "2022-02-01", "2022-03-01","2023-02-01"],
    "dias_mora":[0,0,25,0,11,0,8,7]
})
creditos["Fecha"] = pd.to_datetime(creditos["Fecha"])
creditos = creditos.sort_values(['Cliente', 'Fecha'])
creditos.head()
```

```python
# agrupar los créditos por ID de cliente
grupo_clientes = creditos.groupby('Cliente')
grupo_clientes
```

```python
# calcular la diferencia en días entre cada crédito y el siguiente crédito
diferencia_dias = grupo_clientes['Fecha'].diff().dt.days
diferencia_dias
```

```python
# calcular el promedio de la diferencia en días para cada cliente
promedio_dias = diferencia_dias.groupby(creditos['Cliente']).mean()
promedio_dias
```

```python
# calcular el promedio de la diferencia en días para cada cliente
promedio_dias = diferencia_dias.groupby(creditos['Cliente']).median()
promedio_dias
```
