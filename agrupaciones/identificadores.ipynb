```python
import pandas as pd
```

```python
data = {
    'Fecha': ['2023-05-15', '2023-05-18', '2023-06-02', '2023-06-10', '2023-07-01'],
    'Compra': ['Producto A', 'Producto B', 'Producto C', 'Producto D', 'Producto E'],
    'Monto': [100, 200, 150, 300, 250]
}

df = pd.DataFrame(data)
df['Fecha'] = pd.to_datetime(df['Fecha'])  # Convertir la columna 'Fecha' al formato datetime si no está en ese formato
df
```

```python
df['ID'] = df.groupby(df['Fecha'].dt.to_period('M')).ngroup()
df
```

```python
data = {
    'Cliente': ['Cliente A', 'Cliente A', 'Cliente B', 'Cliente B', 'Cliente C', 'Cliente A',],
    'Fecha': ['2023-05-15', '2023-05-18', '2023-06-02', '2023-06-10', '2023-07-01', '2023-06-15',],
    'Compra': ['Producto A', 'Producto B', 'Producto C', 'Producto D', 'Producto E', 'Producto A',],
    'Monto': [100, 200, 150, 300, 250, 800]
}

df = pd.DataFrame(data)
df['Fecha'] = pd.to_datetime(df['Fecha'])  # Convertir la columna 'Fecha' al formato datetime si no está en ese formato
df
```

```python
df['ID'] = df.groupby([df['Cliente'], df['Fecha'].dt.to_period('M')]).ngroup()
df
```
