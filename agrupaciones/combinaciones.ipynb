```python
import pandas as pd
from itertools import combinations
```

# AIMS 

En este archivo se sacaran las combinaciones posibles por numero

```python
data = {
    'letra': ['A', 'B', 'C', 'D', 'E', 'F'],
    'numero': [['1', '2', '3'],
               ['1', '4', '3'],
               ['5','2', '3', '6'],
               ['5'],
               ['2','6'],
               ['2','6'],
              ]
}
df = pd.DataFrame(data)
df = df.explode('numero').reset_index(drop=True)
df = df.sort_values(by=['letra','numero'], ascending=[False, True])
df
```

```python
df2 = df.groupby('letra')['numero'].agg(lambda x: list(x)).reset_index()
df2
```

```python
df_unique = df2.drop_duplicates(subset='numero').reset_index(drop=True)
df_unique
```

```python
df_unique2 = df_unique.copy()
df_unique2 = df_unique2.loc[:,['numero']]
df_unique2 = df_unique2.reset_index()
df_unique2
```
