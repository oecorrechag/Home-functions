```python
import pandas as pd
import numpy as np
```

```python
df = {'Name': ['A','B','C','D','E','F'], 
      'Age': [10,15,15,12,9,8],
      'Gender': ['F','M','M','M','F',np.nan]
     }
df = pd.DataFrame.from_dict(df)

df['Gender_n'] = df['Gender'].map({'F':1, 'M':0})

df
```

```python
from sklearn.impute import KNNImputer
```

```python
imputer = KNNImputer(n_neighbors = 3)
df_filled = imputer.fit_transform(df[['Age','Gender_n']])
df_filled
```

```python
df_filled2 = pd.DataFrame(df_filled,
                          columns = ['Age','Gender_n'])
df_filled2 = df_filled2.round(0)
df_filled2.head(10)
```
