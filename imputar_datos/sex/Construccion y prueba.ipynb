```python
import pandas as pd
import numpy as np

from list_names import lista_nombres_femeninos, lista_nombres_masculinos
```

# Construccion de la funcion

```python
# Dataset de prueba
df1 = {'Id': [1111,1122,1333,1444,1555], 
       'full_name': ['Marin Moncada Monica Daniela', 'VALENCIA GARCIA CAMILO ANDRES', 'HERRERA SUAZA JUAN CAMILO', 'Rosa H Paris P jupyter', 'ROJAS PEINADO LUIS ANDRES'],
       'sexo':[np.nan, np.nan, np.nan, np.nan, np.nan]
      }
df1 = pd.DataFrame.from_dict(df1)
df1.head()
```

```python
df_aux = df1.copy()
df_name = df_aux.copy()
df_name = df_name.loc[:,['Id', 'full_name']]
lcolumnas = list(df_aux.columns)

df_name['full_name'] = df_name['full_name'].str.title()

nn = df_name['full_name'].str.split(pat = ' ', n = 3, expand = True)

df_name['first_name'],df_name['first_name2'],df_name['last_name'],df_name['last_name2'] = nn[0],nn[1],nn[2],nn[3]

df_name
```

```python
df_name.insert(1, 'sexo_f', np.nan)
df_name.insert(2, 'sexo_m', np.nan)

df_name['sexo_f'] = np.where((df_name['first_name'].isin(lista_nombres_femeninos))
                             | (df_name['first_name2'].isin(lista_nombres_femeninos))
                             | (df_name['last_name'].isin(lista_nombres_femeninos))
                             | (df_name['last_name2'].isin(lista_nombres_femeninos)), 'F', 'N')

df_name['sexo_m'] = np.where((df_name['first_name'].isin(lista_nombres_masculinos))
                             | (df_name['first_name2'].isin(lista_nombres_masculinos))
                             | (df_name['last_name'].isin(lista_nombres_masculinos))
                             | (df_name['last_name2'].isin(lista_nombres_masculinos)), 'M', 'N')

df_name
```

```python
df_aux = df1.copy()
df_name = df_aux.copy()
lcolumnas = list(df_aux.columns)

df_name['full_name'] = df_name['full_name'].str.title()


nn = df_name['full_name'].str.split(pat = ' ', n = 3, expand = True)
df_name['first_name'],df_name['first_name2'],df_name['last_name'],df_name['last_name2'] = nn[0],nn[1],nn[2],nn[3]

df_name.insert(1, 'sexo_f', np.nan)
df_name.insert(2, 'sexo_m', np.nan)

df_name['sexo_f'] = np.where((df_name['first_name'].isin(lista_nombres_femeninos))
                             | (df_name['first_name2'].isin(lista_nombres_femeninos))
                             | (df_name['last_name'].isin(lista_nombres_femeninos))
                             | (df_name['last_name2'].isin(lista_nombres_femeninos)), 'F', 'N')

df_name['sexo_m'] = np.where((df_name['first_name'].isin(lista_nombres_masculinos))
                             | (df_name['first_name2'].isin(lista_nombres_masculinos))
                             | (df_name['last_name'].isin(lista_nombres_masculinos))
                             | (df_name['last_name2'].isin(lista_nombres_masculinos)), 'M', 'N')

conditions = [((df_name['sexo_f'] == 'F') & (df_name['sexo_m'] == 'N')),
              ((df_name['sexo_f'] == 'N') & (df_name['sexo_m'] == 'M')),
              ((df_name['sexo_f'] == 'N') & (df_name['sexo_m'] == 'N'))
             ]
values = ['F','M','N']
df_name['sexo_asignado'] = np.select(conditions, values, default=np.nan)
df_name['sexo_asignado'] = df_name['sexo_asignado'].replace(['N'],np.nan)
df_name = df_name.loc[:,['Id','sexo_asignado']]
 
df_name.head()
```

```python
df_salida = pd.merge(df_aux, df_name, on=['Id'])

df_salida['sexo'] = df_salida['sexo'].fillna(df_salida['sexo_asignado'])

df_salida = df_salida.loc[:, lcolumnas]

df_salida
```
